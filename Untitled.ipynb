```python
# PIPELINE COMPLET DRUGGPT CONDITIONNEL
print("ðŸŽ¯ PIPELINE COMPLET DRUGGPT CONDITIONNEL")
print("="*60)

import torch
import torch.nn as nn
from torch.nn import functional as F
import unidecode
import random
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
import time
import os
from rdkit import Chem
from rdkit.Chem import Descriptors, Crippen, Lipinski
import warnings
warnings.filterwarnings('ignore')

# CONFIGURATION
print("ðŸ”§ CONFIGURATION DU SYSTÃˆME")
print("="*50)

# ParamÃ¨tres du modÃ¨le
batch_size = 32
block_size = 256
learning_rate = 1e-3
n_head = 12
n_embd = 384
n_layer = 12
dropout = 0.15
ff_size_factor = 4
max_iters = 3000
eval_interval = 200
save_interval = 1000

# Device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"ðŸ–¥ï¸  Device: {device}")

# FONCTIONS DE CALCUL DES PROPRIÃ‰TÃ‰S
print("\nðŸ§ª DÃ‰FINITION DES FONCTIONS DE PROPRIÃ‰TÃ‰S")
print("="*50)

def calculate_logp(smiles):
    """Calculate LogP for a SMILES string"""
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            return Crippen.MolLogP(mol)
    except:
        pass
    return None

def count_aromatic_rings(smiles):
    """Count aromatic rings in a molecule"""
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            return Lipinski.NumAromaticRings(mol)
    except:
        pass
    return 0

def count_aliphatic_rings(smiles):
    """Count aliphatic rings in a molecule"""
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            ri = mol.GetRingInfo()
            aromatic_rings = Lipinski.NumAromaticRings(mol)
            return len(ri.AtomRings()) - aromatic_rings
    except:
        pass
    return 0

def has_functional_groups(smiles):
    """Check if molecule has at least one of: -OH, -COOR, -COOH, -NH2"""
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            if mol.HasSubstructMatch(Chem.MolFromSmarts('[OH]')):
                return True
            if mol.HasSubstructMatch(Chem.MolFromSmarts('C(=O)[OX2]')):
                return True
            if mol.HasSubstructMatch(Chem.MolFromSmarts('C(=O)[OH]')):
                return True
            if mol.HasSubstructMatch(Chem.MolFromSmarts('[NH2]')):
                return True
    except:
        pass
    return False

def calculate_r_value(smiles):
    """Calculate R-value (rotatable bonds ratio)"""
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            rotatable_bonds = Lipinski.NumRotatableBonds(mol)
            heavy_atoms = mol.GetNumHeavyAtoms()
            if heavy_atoms > 0:
                return rotatable_bonds / heavy_atoms
    except:
        pass
    return None

def check_single_objective(smiles):
    """Check if molecule satisfies single objective: LogP â‰¤ 3"""
    logp = calculate_logp(smiles)
    return logp is not None and logp <= 3

def check_multiple_objectives(smiles):
    """Check if molecule satisfies multiple structural objectives"""
    aromatic_rings = count_aromatic_rings(smiles)
    aliphatic_rings = count_aliphatic_rings(smiles)
    ring_condition = (aromatic_rings == 2) and (aliphatic_rings == 1)
    functional_group_condition = has_functional_groups(smiles)
    r_value = calculate_r_value(smiles)
    r_condition = (r_value is not None) and (0.05 <= r_value <= 0.50)
    return ring_condition and functional_group_condition and r_condition

print("âœ… Fonctions de propriÃ©tÃ©s dÃ©finies!")

# CHARGEMENT ET PRÃ‰PARATION DES DONNÃ‰ES
print("\nðŸ“ CHARGEMENT DES DONNÃ‰ES")
print("="*50)

def load_and_prepare_data(file_path):
    """Charger et prÃ©parer les donnÃ©es SMILES"""
    try:
        with open(file_path, 'r') as f:
            text = f.read()
        print(f"ðŸ“Š Fichier chargÃ©: {len(text):,} caractÃ¨res")
        return text
    except FileNotFoundError:
        print(f"âŒ Fichier {file_path} non trouvÃ©!")
        return None

# Charger les donnÃ©es (remplacez par votre chemin)
file_path = "s_100_str_+1M.txt"  # Ã€ MODIFIER
text = load_and_prepare_data(file_path)

if text is None:
    # CrÃ©er des donnÃ©es d'exemple si le fichier n'existe pas
    print("ðŸ“ CrÃ©ation de donnÃ©es d'exemple...")
    example_smiles = [
        "CCO", "CCN", "C1=CC=CC=C1", "CC(=O)O", "C1CCCCC1",
        "CN1C=NC2=C1C(=O)N(C(=O)N2C)C", "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O"
    ]
    text = "".join([f"<{smi}>" for smi in example_smiles])
    print(f"ðŸ“Š DonnÃ©es d'exemple crÃ©Ã©es: {len(text)} caractÃ¨res")

# CrÃ©er le vocabulaire
chars = sorted(list(set(text)))
vocab_size = len(chars)
print(f"ðŸ”¤ Vocabulaire: {''.join(chars)}")
print(f"ðŸ“ Taille du vocabulaire: {vocab_size}")

# Encodage/dÃ©codage
stoi = {ch: i for i, ch in enumerate(chars)}
itos = {i: ch for i, ch in enumerate(chars)}
encode = lambda s: [stoi[c] for c in s]
decode = lambda l: ''.join([itos[i] for i in l])

# Encoder les donnÃ©es
data = torch.tensor(encode(text), dtype=torch.long)
n = int(0.9 * len(data))
train_data = data[:n]
val_data = data[n:]

print(f"ðŸ“ˆ DonnÃ©es encodÃ©es: {len(data):,} tokens")
print(f"ðŸ“Š Train: {len(train_data):,}, Validation: {len(val_data):,}")

# ARCHITECTURE DU MODÃˆLE CONDITIONNEL
print("\nðŸ§  ARCHITECTURE DU MODÃˆLE CONDITIONNEL")
print("="*50)

class Head(nn.Module):
    def __init__(self, head_size):
        super().__init__()
        self.key = nn.Linear(n_embd, head_size, bias=False)
        self.query = nn.Linear(n_embd, head_size, bias=False)
        self.value = nn.Linear(n_embd, head_size, bias=False)
        self.register_buffer('tril', torch.tril(torch.ones(block_size, block_size)))
        self.dropout = nn.Dropout(dropout)
        self.head_size = head_size

    def forward(self, x):
        B, T, C = x.shape
        k = self.key(x)
        q = self.query(x)
        v = self.value(x)
        wei = q @ k.transpose(-2, -1) * (self.head_size ** -0.5)
        wei = wei.masked_fill(self.tril[:T, :T] == 0, float('-inf'))
        wei = F.softmax(wei, dim=-1)
        wei = self.dropout(wei)
        return wei @ v

class MultiHeadAttention(nn.Module):
    def __init__(self, num_heads, head_size):
        super().__init__()
        self.heads = nn.ModuleList([Head(head_size) for _ in range(num_heads)])
        self.proj = nn.Linear(n_embd, n_embd)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        out = torch.cat([h(x) for h in self.heads], dim=-1)
        return self.dropout(self.proj(out))

class FeedForward(nn.Module):
    def __init__(self, n_embd):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(n_embd, ff_size_factor * n_embd),
            nn.GELU(),
            nn.Linear(ff_size_factor * n_embd, n_embd),
            nn.Dropout(dropout),
        )

    def forward(self, x):
        return self.net(x)

class Block(nn.Module):
    def __init__(self, n_embd, n_head):
        super().__init__()
        head_size = n_embd // n_head
        self.sa = MultiHeadAttention(n_head, head_size)
        self.ffwd = FeedForward(n_embd)
        self.ln1 = nn.LayerNorm(n_embd)
        self.ln2 = nn.LayerNorm(n_embd)

    def forward(self, x):
        x = x + self.sa(self.ln1(x))
        x = x + self.ffwd(self.ln2(x))
        return x

class ConditionalDrugGPT(nn.Module):
    def __init__(self, n_conditions=2):
        super().__init__()
        self.token_embedding_table = nn.Embedding(vocab_size, n_embd)
        self.position_embedding_table = nn.Embedding(block_size, n_embd)
        self.condition_projection = nn.Linear(n_conditions, n_embd)
        self.blocks = nn.Sequential(*[Block(n_embd, n_head) for _ in range(n_layer)])
        self.ln_f = nn.LayerNorm(n_embd)
        self.lm_head = nn.Linear(n_embd, vocab_size)
        self.apply(self._init_weights)
        
    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)
            if module.bias is not None:
                torch.nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)
            
    def forward(self, idx, conditions=None, targets=None):
        B, T = idx.shape
        tok_emb = self.token_embedding_table(idx)
        pos_emb = self.position_embedding_table(torch.arange(T, device=device))
        x = tok_emb + pos_emb
        
        if conditions is not None:
            cond_emb = self.condition_projection(conditions).unsqueeze(1)
            x = x + cond_emb
        
        x = self.blocks(x)
        x = self.ln_f(x)
        logits = self.lm_head(x)
        
        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape
            logits = logits.view(B*T, C)
            targets = targets.view(B*T)
            loss = F.cross_entropy(logits, targets)
            
        return logits, loss
    
    def generate(self, idx, conditions, max_new_tokens, eos_token, temperature=1.0, top_k=None):
        self.eval()
        with torch.no_grad():
            for _ in range(max_new_tokens):
                idx_cond = idx[:, -block_size:]
                logits, _ = self(idx_cond, conditions)
                logits = logits[:, -1, :] / temperature
                
                if top_k is not None:
                    v, _ = torch.topk(logits, min(top_k, logits.size(-1)))
                    logits[logits < v[:, [-1]]] = -float('Inf')
                
                probs = F.softmax(logits, dim=-1)
                idx_next = torch.multinomial(probs, num_samples=1)
                
                if idx_next.item() == eos_token:
                    break
                    
                idx = torch.cat((idx, idx_next), dim=1)
                
        self.train()
        return idx

print("âœ… Architecture du modÃ¨le dÃ©finie!")

# FONCTIONS D'ENTRAÃŽNEMENT ET GÃ‰NÃ‰RATION
print("\nâš™ï¸  CONFIGURATION DES FONCTIONS D'ENTRAÃŽNEMENT")
print("="*50)

def get_conditional_batch(split):
    data_split = train_data if split == 'train' else val_data
    ix = torch.randint(len(data_split) - block_size, (batch_size,))
    x = torch.stack([data_split[i:i+block_size] for i in ix])
    y = torch.stack([data_split[i+1:i+block_size+1] for i in ix])
    
    conditions = torch.zeros(batch_size, 2)
    for i in range(batch_size):
        choice = random.random()
        if choice < 0.3:
            conditions[i] = torch.tensor([1.0, 0.0])
        elif choice < 0.6:
            conditions[i] = torch.tensor([0.0, 1.0])
        else:
            mix = random.uniform(0.3, 0.7)
            conditions[i] = torch.tensor([mix, 1.0 - mix])
    
    return x.to(device), y.to(device), conditions.to(device)

@torch.no_grad()
def estimate_loss():
    out = {}
    model.eval()
    eval_iters = 100
    
    for split in ['train', 'val']:
        losses = torch.zeros(eval_iters)
        for k in range(eval_iters):
            X, Y, conditions = get_conditional_batch(split)
            logits, loss = model(X, conditions, Y)
            losses[k] = loss.item()
        out[split] = losses.mean()
    
    model.train()
    return out

def generate_conditional_smiles(model, conditions, n_samples=10, temperature=0.8, max_length=100):
    sos_token = encode('<')[0]
    eos_token = encode('>')[0]
    generated_smiles = []
    
    for i in range(n_samples):
        try:
            context = torch.tensor([[sos_token]], dtype=torch.long, device=device)
            if conditions.dim() == 1:
                cond = conditions.unsqueeze(0)
            else:
                cond = conditions[i:i+1] if i < conditions.size(0) else conditions[0:1]
            
            generated = model.generate(
                context, cond, max_new_tokens=max_length,
                eos_token=eos_token, temperature=temperature, top_k=50
            )
            
            smi = decode(generated[0].tolist())
            smi = smi.replace('<', '').replace('>', '').strip()
            
            if smi and Chem.MolFromSmiles(smi) is not None:
                generated_smiles.append(smi)
                
        except:
            continue
            
    return generated_smiles

def evaluate_generated_molecules(smiles_list):
    results = []
    for smi in smiles_list:
        if not smi:
            continue
        try:
            mol = Chem.MolFromSmiles(smi)
            if mol is None:
                continue
            results.append({
                'smiles': smi,
                'logp': calculate_logp(smi),
                'aromatic_rings': count_aromatic_rings(smi),
                'aliphatic_rings': count_aliphatic_rings(smi),
                'has_functional_groups': has_functional_groups(smi),
                'r_value': calculate_r_value(smi),
                'single_objective': check_single_objective(smi),
                'multiple_objectives': check_multiple_objectives(smi),
            })
        except:
            continue
    return pd.DataFrame(results)

print("âœ… Fonctions d'entraÃ®nement configurÃ©es!")

# INITIALISATION ET ENTRAÃŽNEMENT
print("\nðŸš€ INITIALISATION ET ENTRAÃŽNEMENT")
print("="*50)

# CrÃ©er le modÃ¨le
model = ConditionalDrugGPT(n_conditions=2)
model = model.to(device)

# Compter les paramÃ¨tres
total_params = sum(p.numel() for p in model.parameters())
print(f"ðŸ§  ModÃ¨le crÃ©Ã©: {total_params/1e6:.2f}M paramÃ¨tres")

# Optimiseur
optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate, weight_decay=0.01)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=max_iters, eta_min=learning_rate*0.1)

print("âœ… ModÃ¨le et optimiseur initialisÃ©s!")

# FONCTION D'ENTRAÃŽNEMENT PRINCIPALE
def train_conditional_model():
    print("ðŸŽ¯ DÃ‰MARRAGE DE L'ENTRAÃŽNEMENT CONDITIONNEL")
    print(f"â±ï¸  ItÃ©rations: {max_iters}, Batch size: {batch_size}")
    
    losses = []
    train_losses = []
    val_losses = []
    start_time = time.time()
    
    for iter_num in range(max_iters):
        
        if iter_num % eval_interval == 0 or iter_num == max_iters - 1:
            losses_dict = estimate_loss()
            train_loss = losses_dict['train']
            val_loss = losses_dict['val']
            train_losses.append(train_loss)
            val_losses.append(val_loss)
            
            elapsed = time.time() - start_time
            current_lr = optimizer.param_groups[0]['lr']
            progress = (iter_num / max_iters) * 100
            
            print(f"Step {iter_num:4d} ({progress:.0f}%) | "
                  f"Train: {train_loss:.4f} | Val: {val_loss:.4f} | "
                  f"LR: {current_lr:.2e} | Time: {elapsed:.1f}s")
        
        # EntraÃ®nement
        xb, yb, conditions = get_conditional_batch('train')
        logits, loss = model(xb, conditions, yb)
        
        optimizer.zero_grad(set_to_none=True)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        scheduler.step()
        
        losses.append(loss.item())
        
        # Sauvegarde
        if iter_num % save_interval == 0 or iter_num == max_iters - 1:
            checkpoint = {
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'iteration': iter_num,
                'train_losses': train_losses,
                'val_losses': val_losses,
            }
            torch.save(checkpoint, "conditional_drugGPT.pt")
            print(f"ðŸ’¾ Checkpoint sauvegardÃ© itÃ©ration {iter_num}")
    
    total_time = time.time() - start_time
    print(f"\nâœ… ENTRAÃŽNEMENT TERMINÃ‰!")
    print(f"â±ï¸  Temps total: {total_time/60:.1f} minutes")
    print(f"ðŸ“Š Performance: Train={train_losses[-1]:.4f}, Val={val_losses[-1]:.4f}")
    
    # Graphiques
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(losses, alpha=0.7)
    plt.title('Loss par itÃ©ration')
    plt.xlabel('ItÃ©ration')
    plt.ylabel('Loss')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    eval_steps = [i * eval_interval for i in range(len(train_losses))]
    plt.plot(eval_steps, train_losses, 'o-', label='Train')
    plt.plot(eval_steps, val_losses, 's-', label='Validation')
    plt.title('Train vs Validation Loss')
    plt.xlabel('ItÃ©ration')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return losses, train_losses, val_losses

# FONCTIONS DE GÃ‰NÃ‰RATION CIBLÃ‰E
def generate_with_single_objective(n_samples=15):
    conditions = torch.tensor([[1.0, 0.0]], device=device)
    return generate_conditional_smiles(model, conditions, n_samples)

def generate_with_multiple_objectives(n_samples=20):
    conditions = torch.tensor([[0.0, 1.0]], device=device)
    return generate_conditional_smiles(model, conditions, n_samples)

def generate_with_mixed_objectives(n_samples=15):
    conditions = torch.tensor([[0.5, 0.5]], device=device)
    return generate_conditional_smiles(model, conditions, n_samples)

# DÃ‰MONSTRATION
def demonstrate_generation():
    print("\nðŸŽ¯ DÃ‰MONSTRATION DE GÃ‰NÃ‰RATION CONDITIONNELLE")
    print("="*50)
    
    # LogP seulement
    print("\n1. ðŸŽ¯ GÃ‰NÃ‰RATION - OBJECTIF LogP â‰¤ 3")
    molecules = generate_with_single_objective(10)
    df = evaluate_generated_molecules(molecules)
    print(f"   GÃ©nÃ©rÃ©s: {len(df)} molÃ©cules valides")
    if len(df) > 0:
        success = df['single_objective'].sum()
        print(f"   âœ… Satisfont LogP â‰¤ 3: {success}/{len(df)}")
        for _, row in df.head(2).iterrows():
            print(f"      {row['smiles'][:50]}... (LogP: {row['logp']:.2f})")
    
    # Structure seulement
    print("\n2. ðŸ—ï¸  GÃ‰NÃ‰RATION - OBJECTIFS STRUCTURELS")
    molecules = generate_with_multiple_objectives(15)
    df = evaluate_generated_molecules(molecules)
    print(f"   GÃ©nÃ©rÃ©s: {len(df)} molÃ©cules valides")
    if len(df) > 0:
        success = df['multiple_objectives'].sum()
        print(f"   âœ… Satisfont objectifs structurels: {success}/{len(df)}")
        successful = df[df['multiple_objectives']]
        if len(successful) > 0:
            for _, row in successful.head(2).iterrows():
                print(f"      {row['smiles']}")
                print(f"        â†’ Anneaux: {row['aromatic_rings']} arom., {row['aliphatic_rings']} aliph.")
                print(f"        â†’ R-value: {row['r_value']:.3f}")

# LANCEMENT DU PIPELINE
print(f"\nðŸš€ PIPELINE COMPLET PRÃŠT!")
print("="*50)
print("ðŸ“‹ COMMANDES DISPONIBLES:")
print("1. train_conditional_model() - Lancer l'entraÃ®nement")
print("2. demonstrate_generation() - Tester la gÃ©nÃ©ration")
print("3. generate_with_single_objective() - GÃ©nÃ©rer avec LogP contraint")
print("4. generate_with_multiple_objectives() - GÃ©nÃ©rer avec contraintes structurelles")

print(f"\nðŸ’¡ MODIFIEZ file_path LIGNE 125 AVEC VOTRE FICHIER!")
print(f"ðŸ’¡ L'ENTRAÃŽNEMENT PREND 15-20 MINUTES")

# LANCER L'ENTRAÃŽNEMENT (DÃ‰COMMENTER POUR EXÃ‰CUTER)
# losses, train_losses, val_losses = train_conditional_model()

# TESTER LA GÃ‰NÃ‰RATION APRÃˆS ENTRAÃŽNEMENT
# demonstrate_generation()
```

```python
def generate(self, idx, conditions, max_new_tokens, eos_token, temperature=1.0, top_k=None):
    self.eval()
    with torch.no_grad():
        for _ in range(max_new_tokens):
            idx_cond = idx[:, -block_size:]
            logits, _ = self(idx_cond, conditions)
            logits = logits[:, -1, :] / temperature
            
            if top_k is not None:
                v, _ = torch.topk(logits, min(top_k, logits.size(-1)))
                logits[logits < v[:, [-1]]] = -float('Inf')
            
            probs = F.softmax(logits, dim=-1)
            idx_next = torch.multinomial(probs, num_samples=1)
            
            # ArrÃªter si on gÃ©nÃ¨re le token de fin
            if idx_next.item() == eos_token:
                break
                
            idx = torch.cat((idx, idx_next), dim=1)
            
    self.train()
    return idx
```

```python
def generate_conditional_smiles(model, conditions, n_samples=10, temperature=0.8, max_length=100):
    sos_token = encode('<')[0]
    eos_token = encode('>')[0]
    generated_smiles = []
    
    for i in range(n_samples):
        try:
            context = torch.tensor([[sos_token]], dtype=torch.long, device=device)
            if conditions.dim() == 1:
                cond = conditions.unsqueeze(0)
            else:
                cond = conditions[i:i+1] if i < conditions.size(0) else conditions[0:1]
            
            generated = model.generate(
                context, cond, max_new_tokens=max_length,
                eos_token=eos_token, temperature=temperature, top_k=50
            )
            
            # DÃ©coder et nettoyer
            smi = decode(generated[0].tolist())
            # Extraire seulement le contenu entre < et >
            if '<' in smi and '>' in smi:
                start = smi.find('<') + 1
                end = smi.find('>')
                smi = smi[start:end]
            else:
                smi = smi.replace('<', '').replace('>', '')
            
            smi = smi.strip()
            
            # Validation RDKit
            if smi and Chem.MolFromSmiles(smi) is not None:
                generated_smiles.append(smi)
                
        except Exception as e:
            print(f"âŒ Erreur gÃ©nÃ©ration: {e}")
            continue
            
    return generated_smiles
```

```python
def generate_with_multiple_objectives(n_samples=20, max_attempts=50):
    """GÃ©nÃ¨re avec contraintes structurelles avec plusieurs tentatives"""
    conditions = torch.tensor([[0.0, 1.0]], device=device)
    all_valid_molecules = []
    attempts = 0
    
    while len(all_valid_molecules) < n_samples and attempts < max_attempts:
        molecules = generate_conditional_smiles(model, conditions, n_samples=10)
        
        for smi in molecules:
            try:
                # Validation RDKit stricte
                mol = Chem.MolFromSmiles(smi)
                if mol is not None:
                    # VÃ©rifier les objectifs multiples
                    if check_multiple_objectives(smi):
                        all_valid_molecules.append(smi)
                        print(f"âœ… MolÃ©cule valide: {smi}")
            except:
                continue
        
        attempts += 1
        print(f"Tentative {attempts}: {len(all_valid_molecules)}/{n_samples} molÃ©cules valides")
        
        if len(all_valid_molecules) >= n_samples:
            break
    
    return all_valid_molecules[:n_samples]
```

```python
def validate_smiles(smiles):
    """Validation robuste des SMILES"""
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return False
        
        # VÃ©rifier la structure
        smiles_canonical = Chem.MolToSmiles(mol)
        mol2 = Chem.MolFromSmiles(smiles_canonical)
        
        return mol2 is not None
    except:
        return False

def evaluate_generated_molecules(smiles_list):
    """Ã‰valuation amÃ©liorÃ©e avec meilleure gestion des erreurs"""
    results = []
    for smi in smiles_list:
        if not smi or not validate_smiles(smi):
            continue
        try:
            mol = Chem.MolFromSmiles(smi)
            if mol is None:
                continue
                
            results.append({
                'smiles': smi,
                'logp': calculate_logp(smi),
                'aromatic_rings': count_aromatic_rings(smi),
                'aliphatic_rings': count_aliphatic_rings(smi),
                'has_functional_groups': has_functional_groups(smi),
                'r_value': calculate_r_value(smi),
                'single_objective': check_single_objective(smi),
                'multiple_objectives': check_multiple_objectives(smi),
                'mw': Descriptors.MolWt(mol),
                'heavy_atoms': mol.GetNumHeavyAtoms(),
            })
        except Exception as e:
            print(f"Erreur Ã©valuation {smi}: {e}")
            continue
    return pd.DataFrame(results)
```

```python
def test_generation_functions():
    """Test des fonctions de gÃ©nÃ©ration aprÃ¨s correction"""
    print("ðŸ§ª TEST DES FONCTIONS DE GÃ‰NÃ‰RATION")
    print("="*50)
    
    # Test gÃ©nÃ©ration multiple objectifs
    print("ðŸŽ¯ GÃ©nÃ©ration avec objectifs multiples...")
    molecules = generate_with_multiple_objectives(n_samples=5)
    
    if molecules:
        df = evaluate_generated_molecules(molecules)
        print(f"âœ… {len(df)} molÃ©cules valides gÃ©nÃ©rÃ©es")
        
        for _, row in df.iterrows():
            print(f"\nðŸ“‹ {row['smiles']}")
            print(f"   LogP: {row['logp']:.2f}")
            print(f"   Anneaux aromatiques: {row['aromatic_rings']}")
            print(f"   Anneaux aliphatiques: {row['aliphatic_rings']}")
            print(f"   Groupes fonctionnels: {row['has_functional_groups']}")
            print(f"   R-value: {row['r_value']:.3f}")
            print(f"   Objectifs multiples: {row['multiple_objectives']}")
    else:
        print("âŒ Aucune molÃ©cule valide gÃ©nÃ©rÃ©e")
    
    return molecules

# Test immÃ©diat
test_generation_functions()
```
